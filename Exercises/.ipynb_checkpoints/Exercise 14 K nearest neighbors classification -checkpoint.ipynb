### Importing
First lets import all imp modules and libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
digits = load_digits()
dir(digits)
```

We will use 2 Approaches to find best value of **K**, we will first use cross **validation** on a range of **K** and find the optimal value of **K** and then second approach is using **GridSearchCV**

# Approach 1: Cross Validation
We will run a simple for loop that will keep appending the Mean scores in a list after performing the Cross validation on a range of K

```python
k_values = range(1, 21)
mean_accuracies= []
kfold = StratifiedKFold(n_splits=5)
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    crossval = cross_val_score(knn, digits.data, digits.target, cv=kfold)
    mean_accuracies.append(np.mean(crossval))
```

Plot the mean accuracy for each K value to visualize which K gives the best performance.

```python
plt.figure(figsize=(10, 5))
plt.plot(k_values, mean_accuracies, marker='o')
plt.xlabel('Number of Neighbors K')
plt.ylabel('Cross-Validated Accuracy')
plt.title('KNN Classifier Accuracy for Different K Values')
plt.grid()
```

Select the K value that gives the highest cross-validated accuracy.

The np.argmax function is used to find the index of the maximum value in an array. Essentially, it returns the index of the first occurrence of the maximum value along a specified axis.

```python
optimal_k = k_values[np.argmax(mean_accuracies)]
print(f"The optimal value of K is: {optimal_k}")
```

Well, this Approach says Optimal value of K is 2, now lets try **GridSearchCV** approach!

# Approach 2: GridSearchCV

```python
grid = GridSearchCV(KNeighborsClassifier(),
                   {
                       "n_neighbors" : range(1,30)
                   }, cv=kfold, return_train_score=False)
grid.fit(digits.data, digits.target)
grid_df = pd.DataFrame(grid.cv_results_)
grid_df[['param_n_neighbors', 'mean_test_score', 'rank_test_score']]
```

```python
grid.best_params_
```

### As seen from above, even **GridSearchCV** concludes that the most optimal value for K is 2, so we will use 2 and do prediction

# Train Test Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, 
                                                    random_state=2)
```

# KNN Model Training and Prediction

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
predict = knn.predict(X_test)
```

# Confusion Matrix

```python
cm = confusion_matrix(y_test, predict)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
```

# Classification Report

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, predict)
print(report)
```
