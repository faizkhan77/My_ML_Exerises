```python
X = data.drop(['salary_more_then_100k'], axis=1)
y = data['salary_more_then_100k']
```

company and job is Nominal categorical variable meaning we will do OneHotEncoding

degree is an ordinal variable so we will do Label Encoding

### OneHotEncoding

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
onehotencoder_company = OneHotEncoder(sparse=False)
onehotencoder_job = OneHotEncoder(sparse=False)

encoded_company = onehotencoder_company.fit_transform(X[['company']])
encoded_job = onehotencoder_job.fit_transform(X[['job']])

encoded_company_df = pd.DataFrame(encoded_company, columns=onehotencoder_company
                                  .get_feature_names_out(['company'])).astype(int)
encoded_job_df = pd.DataFrame(encoded_job, columns=onehotencoder_job.
                              get_feature_names_out(['job'])).astype(int)
```

```python
X = pd.concat([X, encoded_company_df, encoded_job_df], axis=1)
X
```

### Label Encoding

```python
label_encoder = LabelEncoder()
X['degree'] = label_encoder.fit_transform(X['degree'])
X
```

### Drop Unnecessary columns

```python
X.drop(['company', 'job', 'company_google', 'job_sales executive'], axis=1, inplace=True)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
