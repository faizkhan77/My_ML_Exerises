# SKLearn OneHotEncoder

```python
label_encoder = LabelEncoder()
label_encoded_data = data.copy()
label_encoded_data
```

```python
label_encoded_data.town = label_encoder.fit_transform(label_encoded_data.town)
label_encoded_data
```

```python
# using values cuz we want a 2D array not a dataframe
X = label_encoded_data.drop(['price'], axis=1).values 
y = data.price
y
```

```python
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder='passthrough')
ct
```

```python
X = X[:, 1:]
X
```

```python
reg2 = LinearRegression()
reg2.fit(X, y)
```

```python
reg2.predict([[0, 1, 3300]])
```

```python
reg2.predict([[1, 0, 2800]])
```

# OneHotEncoder but using ChatGPT Steps

```python
data = pd.read_csv("C:\\Users\\User\\OneDrive\\Documents\\Faiz Khan Program\\Machine Learning\\ML with codebasics\\homeprices3.csv")
data
```

### 1. Instantiate OneHotEncoder:

sparse=False ensures the output is a dense array

```python
encoder = OneHotEncoder(sparse=False)
```

### 2. Fit and Transform Data:

```python
encoder_Data = encoder.fit_transform(data[['town']])
encoder_Data
```

### 3.Convert to DataFrame:

get_feature_names_out: Generates the names for the new binary columns.

```python
encoded_df = pd.DataFrame(encoder_Data, columns=encoder.get_feature_names_out(['town'])).astype(int)
encoded_df
```

### 4. Concatenate Original DataFrame with Encoded Data:

```python
data = pd.concat([data, encoded_df], axis=1)
data
```

### 5. Drop the Original Categorical Column (Optional):

```python
data.drop(['town', 'town_west windsor'], axis=1, inplace=True)
```

```python
data
```

### 6. Split input and output and Train your model

```python
X = data.drop('price', axis=1)
y = data.price
```

```python
reg3 = LinearRegression()
reg3.fit(X, y)
```

```python
reg3.predict([[3300, 0, 0]])
```

```python
reg3.predict([[2800, 0,1]])
```
