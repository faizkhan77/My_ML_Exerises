# IQR Rule for Outliers Detection

```python
# First sort in Ascending order
sorted_val = df['Insulin'].sort_values()

# Then Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = sorted_val.quantile(0.25)
Q3 = sorted_val.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(upper_bound)
print(lower_bound)
```

```python
df[ (df['Insulin'] < lower_bound) | (df['Insulin'] > upper_bound) ]
```

```python
# Plot histogram
plt.hist(df['Insulin'], bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
```

```python
import numpy as np
import seaborn as sns
from scipy import stats  
# Plot histogram and density plot
plt.figure(figsize=(10, 6))

# Plot histogram
sns.histplot(df['Insulin'], bins=10, kde=True, color='g', edgecolor='black', alpha=0.6, stat='density')

# Plot normal distribution curve
mean_insulin = np.mean(df['Insulin'])
std_insulin = np.std(df['Insulin'])
x = np.linspace(min(df['Insulin']), max(df['Insulin']), 100)
y = stats.norm.pdf(x, mean_insulin, std_insulin)
plt.plot(x, y, color='blue', linewidth=2, label='Normal Distribution')

# Add labels and title
plt.xlabel('Insulin Levels')
plt.ylabel('Density')
plt.title('Density Plot with Normal Distribution Overlay')
plt.legend()
```

# Train Test split
The stratify argument in the train_test_split method is used in the context of splitting a dataset into training and testing sets. When you set stratify to a certain variable (usually the target variable), it ensures that the splitting process maintains the same proportion of that variable's classes in both the training and testing sets.

For instance, if you have a dataset with a target variable that represents classes like "Yes" and "No" or "Class A," "Class B," and "Class C," using stratify ensures that when you split the data, both the training and testing sets will have similar proportions of each class as the original dataset.

This is particularly useful in classification problems where you want to maintain the representation of different classes in both sets, ensuring that the model learns from a balanced sample of the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, stratify=y,
                                                    random_state=10)
```

# Model
Now, we will use a Decision Tree Classifier to Train a Standalone model but you can use SVM, KNN or any Classification model

We are using Decision Tree to demonstrate that Decision Tree is relatively imbalance Classifier, it can Overfit and might generate an High Variance model so lets Train Decision Tree first and we will use cross_val_score here to just try it out on different dataset rather than just X_train or X_test

# Standalone model (Decision Tree)

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

np.mean(cross_val_score(DecisionTreeClassifier(), X, y, cv=5))
```

# Bagging Classifier
You can read all the Arguments in **Bagging Classifier** but we will just use couple of Arguments such as 
1. base_estimator : Meaning which estimator ur using so here will pass **Decision Tree**
2. n_estimator : Meaning how many Subsets of Datasets, or basically how many models cuz each Subset will be Trained on each model, so we will do 100 subsets but u can do Trial and Error and figure out which one is giving u the best performance
3. max_samples : Meaning how many samples, like in above Theoratical part we used for example 70 samples out of 100, here can say **0.8** which basically means use 80%
4. oob_score : **OOB** means Out of Bag, when you are sampling, cuz u are doing Random sampling, by the law of probability, lets say u have 100 samples, when u do subsets, u are not going to cover all the 100 samples, lets say in all the subsets, 29 samples didnt appear at all so all the models Trained on each subsets hasnt seen the data 29, so u can use that 29 to Test the accuracy of these models, so ur kind of treating these 29 samples as a Test dataset. <br><br>In a bagging classifier, we make many "copies" of our original data by picking random parts of it. Some data doesn't get picked in each "copy." This leftover data is like a test to check how well each copy can predict things it hasn't seen before.<br><br>We use this leftover data to see how good each copy is at predicting. Then we combine all these results to understand how accurate our whole bunch of copies might be when dealing with new, unseen stuff. This leftover data helps us estimate how well our bagging method will work on new things without needing extra test data.<br><br>
The OOB score is the average accuracy of each model when it's tested only on the samples it didn't see during training. When you combine all these individual scores together, you get an estimate of how accurate your overall model might be when faced with new, unseen data. This score helps you understand how well your group of models might perform without needing a separate test set.

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
bag_model.fit(X_train, y_train)
```

Check oob score as below

```python
bag_model.oob_score_
```

So do u realize, that we didnt even have to use X_test or y_test since OOB did the Test job for u, but we can also do regular scoring using X_test and y_test as below

```python
bag_model.score(X_test, y_test)
```

So u can see we are getting around 75-78% score using Bagging as compared to using Standalone model above we only got like 68%, so as u could see there is an Improvement

We can also use cross_val_score to check as below

```python
np.mean(cross_val_score(bag_model, X, y, cv=5))
```

So again we got a pretty good score, hence we can see, Standalone model gave us 68% accuracy and the Bagged model gave us around 77% which is defo an improvement

So for unstable classifier like Decision Tree, Bagging helps.

Sometimes u might have unstable classifier like Decision Tree and sometimes ur Datasets is such that theres so many NULL values and ur columns are such that ur resulting model have High Variance so it make sense to use Bagging Classifier
