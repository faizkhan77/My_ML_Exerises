# MANUALLY CREATING SVM

## Import useful libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Create X then define the DataFrame

```python
# Define the dataset
X = np.array([[3, 65],
              [5, 80],
              [2, 50],
              [7, 90],
              [4, 75],
              [6, 85],
              [1, 40],
              [8, 95]])
```

```python
X
```

## Adding Passem sem 6 and label column

```python
df = pd.DataFrame(X, columns=['Num of Hours studies', 'Sem 5 marks'])
df['Passed sem 6'] = [0, 1, 0, 1, 1, 1, 0, 1]
df
```

```python
df['label'] = df['Passed sem 6'].apply(lambda x : -1 if x == 0 else x)
df
```

## Creating SVM manually and finding the optimal value for w and b using Gradient Descent Algorithm

```python
def train_svm(X, y, learning_rate=0.01, iteration=1000):
    # Initialize weights and bias
    w = np.zeros(df.iloc[:, :2].shape[1])
    b = 0
    
    for i in range(iteration):
        for index, item in enumerate(X):
            y_predict = np.dot(item, w) + b
            
            if y[index] * y_predict <= 1:
                w += learning_rate * (item * y[index])
                b += learning_rate *  y[index]
    
    return w, b
    
```

1. Learning rate: It determines how much we adjust the weights and bias during each iteration of training. It's like the step size we take while trying to find the optimal solution. A higher learning rate means we take larger steps, but it may lead to overshooting the optimal solution. A smaller learning rate means we take smaller steps but may take longer to converge.

2. Epochs/iteration: It defines the number of times we iterate through the entire dataset during training. Each iteration over the entire dataset is called an epoch. Increasing the number of epochs allows the model to see the data more times and potentially learn better representations. However, too many epochs may lead to overfitting, where the model learns noise from the training data.

3. prediction = np.dot(X[i], w) + b: This line computes the prediction for the current data point (X[i]) by taking the dot product between the feature vector (X[i]) and the weight vector (w), and then adding the bias term (b). Essentially, it calculates the linear combination of features weighted by the learned coefficients.

4. if y[i] * prediction <= 1:: This line checks if the prediction for the current data point is on or inside the margin. If the product of the true label (y[i]) and the prediction is less than or equal to 1, it means the data point is either misclassified or is within the margin. In SVM, we want to maximize the margin, so we update the weights and bias only for misclassified points or points within the margin.

5. w += learning_rate * (X[i] * y[i]): This line updates the weights vector (w) using the gradient descent algorithm. It adjusts the weights proportional to the learning rate (learning_rate) and the misclassified feature vector (X[i]) multiplied by its true label (y[i]). This update rule tries to minimize the classification error by moving the decision boundary in the direction that reduces the error.

6. w += learning_rate * (X[i] * y[i]): This line updates the weights vector (w) using the gradient descent algorithm. It adjusts the weights proportional to the learning rate (learning_rate) and the misclassified feature vector (X[i]) multiplied by its true label (y[i]). This update rule tries to minimize the classification error by moving the decision boundary in the direction that reduces the error.

7. b += learning_rate * y[i]: Similarly, this line updates the bias term (b) by adding the product of the learning rate and the true label (y[i]). The bias term helps in shifting the decision boundary up or down.

### Split X and y and Training the model

```python
X = df.drop(['label', 'Passed sem 6'], axis=1).values
y = df['label'].values
w, b = train_svm(X, y)

print(f"Weights: {w}, Bias: {b}")
```

### After training, we can visualize the decision boundary and the data points:

```python
# Visualize the data points and decision boundary
plt.scatter(df['Num of Hours studies'], df['Sem 5 marks'], c=y, cmap='bwr', marker='o', 
            edgecolors='k', label='Data points')

# Plot the decision boundary
x_values = np.array([np.min(X[:, 0]), np.max(X[:, 0])])
y_values = (-1 / w[1]) * (w[0] * x_values + b)
plt.plot(x_values, y_values, label='Decision boundary')

plt.xlabel('Number of hours studied')
plt.ylabel('Semester 5 marks')
plt.title('SVM Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```
